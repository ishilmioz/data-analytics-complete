1. Python program to calculate the factorial of a number.

```python
def factorial (n):
    if n == 0:
        return 1
    else:
        return n * factorial (n-1)
number = int(input("Enter a number to calculate the factorial: "))
print(f"Result: {factorial(number)}")
```

2. Python program to check if a variable is of a specific data type.

```python
var = "Ismail"

if isinstance (var, str):
    print(f"{var} is a string.")
else:
    print(f"{var} is not a string")
```

4. Python program to sort a list of numbers in ascending order.

```python
numbers = [int(a) for a in input("Enter numbers with spaces.").split()]
numbers.sort()

print(f"sorted list: {numbers}" )
```

5. Python program that asks the user to input a number and prints whether the number is positive and even, positive and odd, or negative.
(Nested if statement)

```python
num = int(input("Enter a number"))

if num > 0:
    if num % 2 == 0:
        print(f"{num} is positive and even.")
    else:
        print(f"{num} is positive and odd.")
elif num < 0:
    print(f"{num} is negative.")
else:
    print(f"{num} is zero.")        
```

7. Python program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.


```python
sum = 0

while True:
    num = int(input("Enter number: (0 to stop:)"))
    sum = sum + num
    if num == 0:
        break
print(f"sum of the numbers: {sum}")
```

8. Python program that calculates the factorial of a number input by the user using a while loop.

```python
num = int(input("Enter a number to calculate the factorial: "))
factorial = 1
i = 1

while i <= num:
    factorial = factorial * i
    i = i + 1
print(f"Factorial: {factorial}")
```

9. Python program that calculates the sum of the digits of a number input by the user using a while loop.

```python
num = int(input("Enter a number: "))
sum_of_digits = 0

while num > 0:
    digit = num % 10
    sum_of_digits = sum_of_digits + digit
    num = num // 10
print(f"sum: {sum_of_digits}")
```

10. Python program that checks if a number input by the user is a prime number using a for loop.

```python
num = int(input("Enter a number"))
is_Prime = True

if num <= 1:
    is_Prime = False
else:
    for i in range(2,int(num ** 0.5)+1):
        if num % i == 0:
            is_Prime = False
if is_Prime:
    print(f"{num} number is a prime number.")
else:
    print(f"{num} number is not a prime number.")
         
```

11. Python program that prints the first n Fibonacci numbers, where n is input by the user.

```python
n = int(input("Enter a number"))
a , b = 0 , 1
count = 0

while count<n:
    print(a)
    a , b = b , b+a
    count = count + 1
```
